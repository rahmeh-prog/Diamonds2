#### Transformation Pipelines

```python
diamonds.info()
```

###### numerical pipeline

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

diamonds_num=diamonds.drop(['cut','clarity','color'],axis=1)
```

```python
from sklearn.impute import SimpleImputer
num_pipeline = Pipeline([
 ('imputer', SimpleImputer(strategy="median")),
 ('std_scaler', StandardScaler()),
 ])
diamonds_num_tr = num_pipeline.fit_transform(diamonds_num)
```

```python
diamonds_p= pd.DataFrame(diamonds_num_tr,columns=['carat','depth','table','price','x','y','z'],index=diamonds.index)
diamonds_p.describe()
```

###### full pipeline

```python
from sklearn.compose import ColumnTransformer
num_attribs = list(diamonds_num)
cat_attribs = ["cut","clarity","color"]
full_pipeline = ColumnTransformer([
 ("num", num_pipeline, num_attribs),
 ("cat", OrdinalEncoder(), cat_attribs),
 ])
diamonds_prepared = full_pipeline.fit_transform(diamonds)
```
